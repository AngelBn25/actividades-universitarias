```python
frutas=["manzana","banano","pera","uva","naranja"]
#define el ultimo valor de la lista
uf=len(frutas)-1
#impirime el primer y ultmo valor
print("el primer elemento de la lista es: ",frutas[0]," y el ultimo elemento de la lista es: ",frutas[uf])
#añade mango a a lista
frutas.append("mango")
#inicia un bucle para verificar parte por parte hasta eliminar pera de la lista
for i in frutas:
    if i == "pera":
      frutas.remove(i)
      print("se elimino la pera de la lista")
      break
    else:
        print("no se encuentra pera en la lista")
```

```python
# 2. Lista de numeros del 1 al 10

numeros = list(range(1, 11))

print("Suma total:", sum(numeros))
print("Numero mas grande:", max(numeros))
print("Numero mas pequeño:", min(numeros))
print("Promedio:", sum(numeros) / len(numeros))
```

```python
# 4. Lista de edades

edades = [12, 17, 18, 20, 15, 22, 13]

# Crear lista de mayores de edad (>_18)
mayores = [edad for edad in edades if edad >= 18]

# Imprimir cantidad y lista resultante de los mayores
print("Cantidad de mayores de edad:", len(mayores))
print("Lista de mayores de edad:", mayores)
```

```python
# 5. Lista de notas

notas = [3.5, 4.2, 2.8, 3.0, 4.0]

# Calcular el promedio
promedio = sum(notas) / len(notas)

# Imprimir el promedio
print("El promedio de las notas es:", promedio)

# Verificar si es aprobado o reprobado
if promedio >= 3.0:
    print("El alumno esta aprobado.")
else:
    print("El alumno esta reprobado.")
# Mostrar notas ordenadas
notas.sort()
print("Notas ordenadas:", notas)
```

```python
# 6. Programa de productos y precios

productos = []
precios = []

# Ingresar los datos
for i in range(5):
    nombre = input(f"Ingrese el nombre del producto {i + 1}: ")
    precio = float(input(f"Ingrese el precio del producto { i + 1}: "))
    productos.append(nombre)
    precios.append(precio)

# Calcular el total
total = sum(precios)

# Imprimir el total
print("\nPrecio total de los productos:", total)

# Productos mas caro y mas barato
indice_caro = precios.index(max(precios))
indice_barato = precios.index(min(precios))
print("Producto mas barato:", productos[indice_barato])
print("Producto mas caro:", productos[indice_caro])
```
