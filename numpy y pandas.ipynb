```python
import numpy as np
import pandas as pd

lista=[]
lista = np.random.randint(1, 101, size=10).tolist()
print(lista)
random_numbers_series = pd.Series(lista)
random_numbers_series.name = 'numeros aletorias'
random_numbers_series.index.name = 'idx'
print("\n--- Pandas Series with Names ---")
print(random_numbers_series)
print(random_numbers_series.index)
squared_lista = (np.array(lista)**2).tolist()
print(f"Lista al cuadrado: {squared_lista}")
```
